## Collation des matériaux du dossier « Écureuil », avec l'outil CollateX

```python
from collatex import *

collation = Collation()

temoin_JournalA = open( "textes/PresTombe/Pres_1938Journal.txt", encoding='utf-8' ).read()
temoin_JournalB = open( "textes/PresTombe/Pres_1940SuisseRomande.txt", encoding='utf-8' ).read()


collation.add_plain_witness( "Journal", temoin_JournalA )
collation.add_plain_witness( "SuisseR", temoin_JournalB )

table = collate(collation, segmentation=False, layout='vertical', output='html2')
print(table)
```
